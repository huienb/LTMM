# LÝ THUYẾT MẬT MÃ
## 1. Cơ sở toán học
### 1.1. Số học các số nguyên 
Tập số nguyên Z, tập số nguyên dương Z+ đóng kín đối với phép cộng, trừ, nhân nhưng không đóng kín với phép chia.  

VD: a = 2, b = 3 thuộc Z 
 - a + b = 5 thuộc Z
 - a - b = -1 thuộc Z
 - a * b = 6 thuộc Z
 - a / b = 2/3 không thuộc Z
 
Hai số nguyên bất kỳ a và n, n>1:  
$$a = q.n + r, 0 < r < n$$   
 - q = a div n: số chia
 - r = a mod n: số dư

Ước chung lớn nhất:  
Số nguyên d được gọi là ước số chung lớn nhất của a và b nếu d > 0, d là ước số chung của a và b, và mọi ước số chung của a và b đều là ước số của d.  
$$d = gcd(a,b)$$

Cách tìm ước chung lớn nhất: Sử dụng thuật toán Euclidean 
![Thuật toán Euclidean](.\euclidean.png)

Thuật toán Euclidean mở rộng:  
Cho 2 số nguyên a, b, tìm 2 số nguyên s và t sao cho  
$$s * a + t * b = gcd(a,b)$$
![Process](.\extended_euclidean_process.png)  
![Algorithm](.\extended_euclidean_algo.png)

Số nguyên tố: Là một số nguyên lớn hơn 1 chỉ có 2 ước số là 1 và chính nó.  
Hai số a và b được gọi là nguyên tố với nhau nếu gcd (a,b) = 1.  
Một số nguyên lớn hơn 1 bất kỳ đều có thể khai triển dưới dạng tích của các số nguyên tố. Dạng khai triển như vậy gọi là dạng khai triển chính tắc của số đó.  

### 1.2. Đồng dư
Hai số nguyên a và b đồng dư với nhau theo mod n khi (a - b) chia hết cho n. Hai số nguyên thuộc cùng một lớp tương đương khi và chỉ khi chúng cho cùng một số dư nếu chia cho n.  

Mỗi lớp tương đương được đại diện bởi một số duy nhất trong tập hợp Zn = {0, 1, 2, ..., n-1} là số dư chung khi chia các số trong lớp đó cho n.  
Ví dụ: Z2 = {0, 1}

![Z và Zn](.\Z_Zn.png)

### 1.3. Các phép toán trong tập Zn
Ta xét đến 3 phép toán hai ngôi trong tập Z có thể được định nghĩa trong tập Zn.    
* (a + b) mod n = c
* (a - b) mod n = c
* (a * b) mod n = c

Các tính chất:
 - (a + b) mod n = [(a mod n) + (b mod n)] mod n
 - (a - b) mod n = [(a mod n) - (b mod n)] mod n
 - (a * b) mod n = [(a mod n) * (b mod n)] mod n
 
Nghịch đảo cộng:  
Cho a thuộc Zn. Một số nguyên b thuộc Zn được gọi là nghịch đảo cộng của a theo mod n nếu (a + b) đồng dư với 0 theo mod n. 

Nghịch đảo nhân:  
Cho a thuộc Zn. Một số nguyên x thuộc Zn được gọi là nghịch đảo nhân của a theo mod n nếu (a.x) đồng dư với 1 theo mod n. Tồn tại số x như vậy thì ta nói a khả nghịch.  
Sử dụng thuật toán Euclidean mở rộng để tìm nghịch đảo nhân.

### 1.4. Thặng dư thu gọn và phần tử nguyên thủy
Tập $Z_n$ = {0, 1, 2, ..., n-1} thường được gọi là tập thặng dư đầy đủ theo mod n, vì mọi số nguyên bất kỳ đều có thể tìm được trong $Z_n$ một số đồng dư theo mod n.  
Tập các thặng dư thu gọn theo mod n được định nghĩa là tập $Z_n^*$ = {a thuộc $Z_n$: gcd(a, n) = 1}, tức $Z_n^*$ là tập con của $Z_n$ bao gồm tất cả các phần tử nguyên tố với n.  


## 2. Các hệ mật cổ điển
### 2.1. Hệ mật Caesar
Đây là hệ mật thay thế được biết đến sớm nhất và đơn giản nhất, được tạo ra bởi Julius Caesar. Hệ mật Caesar thực hiện thay thế mỗi chữ cái trong bản rõ bằng một chữ cái cách đó 3 vị trí trong bảng chữ cái để có được bản mật. Bảng chữ cái quay vòng, vì vậy chữ cái sau Z là A.  

Xét trong bảng chữ cái tiếng Anh gồm có 26 chữ cái. Đánh số các chữ cái từ 0 đến 25. Khi đó thuật toán mã hóa Caesar có thể được biểu diễn như sau:
$$C = E(3, P) = (P + 3) mod 26$$
$$p = D(3, C) = (C - 3) mod 26$$

Tổng quát: 
$$C = E(k, P) = (P + k) mod 26$$
$$P = D(k, C) = (C - k) mod 26$$
trong đó k là giá trị từ 1 đến 25.

Ví dụ: Hệ mật Caesar với k = 15  
plaintext:   hello  
ciphertext:  WTAAD

Khi biết một bản mật là của hệ mật Caesar, thám mã chỉ cần thử 25 khóa có thể xảy ra đến khi tìm được một bản tin có nghĩa.

Để mở rộng không gian khóa, người ta sử dụng hoán vị các vị trí trong bảng chữ cái. Như vậy không gian khóa mở rộng từ 25 khóa lên $26!$ khóa. 

Tuy nhiên hệ mật này vẫn dễ bị phá dựa vào tần suất xuất hiện của các chữ cái. Đầu tiên, tìm tần suất xuất hiện của các chữ cái trong bản tin mật và so sánh với phân bố chuẩn tần suất xuất hiện của các chữ cái trong bảng chữ cái, tìm ra các cặp có tần suất xuất hiện tương đồng.

### 2.2. Hệ mật Affine
Hệ mật Affine sử dụng một cặp khóa, trong đó k1 thuộc $Z_26^*$ và k2 thuộc $Z_26$.
$$C = (P * k1 + k2) mod 26$$
$$P = ((C - k2)* $k1^-1$) mod 26$$

Hệ mật Caesar là một trường hợp đơn giản của hệ mật Affine với k1 = 1

Ví dụ: Hệ mật Affine với k = (7, 2)  
plaintext:   hello  
ciphertext:  ZEBBW

### 2.3. Hệ mật Vigenere
Hệ mật Vigenere là một phương pháp mã hóa bằng cách sử dụng xen kẽ một số phép mật mã hóa Caesar khác nhau dựa trên các chữ cái trong khóa. Đây là một trong những hệ mật nổi tiếng nhất và đơn giản nhất trong những hệ mật thay thế đa ký tự.
Hệ mật được biểu diễn như sau:
$$P = P_0,P_1,P_2,...,P_(n-1)$$
$$K = k_0,k_1,k_2,...,k_(m-1), thường m<n$$
$$C = C_0,C_1,C_2,...,C_(n-1)$$
trong đó: $C_i = (P_i + k_i) mod 26$  
Ta có độ dài của khóa thường nhỏ hơn độ dài của bản rõ, vì vậy khi mật mã hóa, khóa sẽ được lặp lại cho đến khi bằng độ dài của bản rõ. Như vậy, ta có: 
$$C_i = (P_i + k_(i mod m)) mod 26$$

Có thể sử dụng bảng sau để mật mã hóa và giải mật mã hóa của hệ mật Vigenere như sau:  
- Khi mật mã hóa, sử dụng mỗi chữ cái trong bản rõ và chữ cái tương ứng trong khóa như là chỉ số cột và hàng, từ đó có thể tìm ra chữ cái tương ứng cho bản mật
- Khi giải mã, quá trình làm ngược lại

## 3. Hệ mật DES
Hệ mật DES (Data Encryption Standard) là một hệ mật khóa đối xứng được công bố bởi the National Institue of Standards and technology NIST (Viện tiêu chuẩn và công nghệ quốc gia). Đây là một phương pháp mật mã hóa được FIPS (Tiêu chuẩn xử lý thông tin Liên bang Hoa Kỳ) chọn làm tiêu chuaarn chính thức vào năm 1976. Sau đó hệ mật này được sử dụng rộng rãi trên toàn thế giới.  

Hiện nay, DES không còn đủ an toàn cho nhiều ứng dụng (vì độ dài khóa 56 bit là quá nhỏ). Thay vào đó, thuật toán được sử dụng là Triple DES (3DES) và dần được thay thế bằng AES (Advanced Encryption Standard).

DES là mật mã khối với đầu vào là các khối có độ dài 64 bit, vào tạo ra 64 bit bản mật, sử dụng 56 bit khóa. Cùng một thuật toán và khóa được sử dụng cho quá trình mật mã hóa và giải mật mã hóa.
![](.\des_block.png)

Trong thực tế, khóa ban đầu của hệ mật DES có độ dài 64 bit. Tuy nhiên, trước khi mật mã hóa, mỗi bit thứ 8 của khóa sẽ bị loại bỏ, tạo thành khóa có độ dài 56 bit. Các bit bị loại bỏ bao gồm bit số 8, 16, 24, 32, 40, 48, 56.  

Các bước mật mã hóa DES được thực hiện như sau:
1. Khối bản rõ 64 bit được đưa vào initial Permuation (IP) function (hàm hoán vị ban đầu).
2. Bản rõ được hoán vị. Tiến hành hàm DES
3. Chia bản đã được hoán vị thành 2 nửa: nửa trái và nửa phải.
4. Mỗi nửa trái và phải đi qua 16 vòng mật mã hóa
5. Mỗi nửa trái và phải được hợp lại và đi qua hàm hoán vị cuối cùng (Final Permutation)
6. Thu được khối bản mật 64 bit
![](.\des_general.png)

IP và FP hoán vị các bit theo bảng sau:
![](.\des_ip_fp.png)

Sau khi 64 bit bản rõ được hoán bị trong IP sẽ được chia làm 2 nửa. Mỗi nửa này sẽ đi qua 16 vòng, mỗi vòng bao gồm các bước: Tạo khóa, Expansion P-box, XOR, S-box, Straight P=box, và đổi chỗ.  
1. Tạo khóa: Từ khóa 56 bit ban đầu, các khóa con 48 bit khác nhau sẽ được tạo ra trong mỗi vòng. Với mỗi vòng, các bit của khóa 56 bit dịch trái 1-2 bit, tùy thuộc vào số vòng. Sau đó, lấy ra 48 bit từ khóa 56 bit đã dịch, tạo thành các khóa con.
![](.\des_bit_shift.png)
![Key - compression](.\des_key_48.png)
2. Expansion P-box: Mỗi nửa 32 bit bản rõ sẽ được mở rộng thành 48 bit theo bảng sau:
![](.\ex_pbox.png)
3. XOR (làm trắng): Sau bước expansion P-box, DES sử dụng XOR cho 48 bit vừa mở rộng và khóa con đã tạo ở bước trước.
4. S-box: Sử dụng 8 S-box với 6 bit đầu vào và 4 bit đầu ra, thu được kết quả là 32 bit
5. Straight P-box: thực hiện hoán vị các bit theo bảng
![](.\s_pbox.png)
